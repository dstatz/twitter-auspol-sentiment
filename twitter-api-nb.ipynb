```python
import os
import requests
import base64
import pandas as pd
import numpy as np
import credentials
```

```python
# grab keys from credentials file

API_KEY = credentials.creds['API_KEY']
API_SECRET = credentials.creds['API_SECRET']

# combine api key and secret and base64 encode

key_secret = f"{API_KEY}:{API_SECRET}".encode('ascii')
b64_encoded_key = base64.b64encode(key_secret)
b64_encoded_key = b64_encoded_key.decode('ascii')

# set up request from twitter api

base_url = 'https://api.twitter.com/'
auth_url = f'{base_url}oauth2/token'

auth_headers = {
    'Authorization': f'Basic {b64_encoded_key}',
    'Content-Type': 'application/x-www-form-urlencoded;charset=UTF-8'
}

auth_data = {
    'grant_type': 'client_credentials'
}

auth_resp = requests.post(auth_url, headers=auth_headers, data=auth_data)

print(f"API Request Status Code: {auth_resp.status_code}")

access_token = auth_resp.json()['access_token']
```

```python
# tweets search

search_headers = {
    'Authorization': f'Bearer {access_token}'
}


# %23 is URL encoded form of #

search_params = {
    'q': '%23auspol',
    'result_type': 'mixed',
    'count': 100,
    'include_entities': 'true'
}

search_url = f"{base_url}1.1/search/tweets.json"

search_resp = requests.get(search_url, headers=search_headers, params=search_params)
```

```python
for tweet in search_resp.json()['statuses']:
    print(tweet)
    print('\n')
```
